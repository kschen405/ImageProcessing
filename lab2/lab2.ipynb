```python
from canny import Canny, convolve
import os
import glob
import cv2
from matplotlib import pyplot as plt

DIR = "data/"
img_lst = []
for filename in glob.glob(DIR + "*.jpg"):
    img = cv2.imread(filename)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    img_lst.append(img)
    print(filename)
    plt.imshow(img, cmap='gray')
```

```python
canny_detector = Canny(img_lst, threshold1= 240, threshold2= 200, gaussian_sigma=3, gaussian_k_size=5 )
# output_lst = canny_detector.detect_edge()
output_lst = canny_detector.gaussian_blur(img, 5, 1.5)
```

```python
plt.figure(figsize=(20, 40))
for i, img in enumerate(output_lst):
    if img.shape[0] == 3:
        img = img.transpose(1,2,0)
        print("img.shape = ", img.shape)
    plt_idx = i+1
    plt.subplot(2, 2, plt_idx)
    plt.imshow(img, cmap='gray')
plt.show()
```
